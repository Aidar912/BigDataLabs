<h1 style="color:green;font-size:19px">Построить график произвольной квадратичной функции, масштаб задать произвольно. Сделать настройки подписей осей, заголовка фигуры</h1>


```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-10, 10, 100)
y = x ** 2 + 3 * x + 2

fig, ax = plt.subplots(figsize=(8, 6))
plt.plot(x, y, label='y = x^2 + 3x + 2', color='cyan', linewidth=2)

plt.xlabel('X', ha='center')
plt.ylabel('Y', va='center', rotation=0, ha='center')
plt.title('График квадратичной функции')

plt.legend()

ax.spines['left'].set_position('center')
ax.spines['bottom'].set_position('center')
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')

plt.grid(True)
plt.show()
```


<h1 style="color:green;font-size:19px">Вывести графики п.2 и 3 в разные поля при помощи функции subplot()</h1>


```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-10, 10, 100)
y1 = x ** 2 + 3 * x + 2
y2 = 2 * np.cos(5 * x + 3)

plt.figure(figsize=(12, 6))

ax = plt.subplot(1, 2, 1)
plt.plot(x, y1, label='y = x^3 - 2x^2 + x + 6', color='cyan', linewidth=2)
plt.xlabel('X', ha='center')
plt.ylabel('Y', va='center', rotation=0, ha='center')
plt.title('График функции y = x^3 - 2x^2 + x + 6')
plt.legend()
plt.grid(True)

ax.spines['left'].set_position('center')
ax.spines['bottom'].set_position('zero')
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')

plt.subplot(1, 2, 2)
plt.plot(x, y2, label='y = 2 * cos(5x + 3)', color='r', linewidth=2)
plt.xlabel('X', ha='center')
plt.ylabel('Y', va='center', rotation=0, ha='center')
plt.title('График функции y = 2 * cos(5x + 3)')
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()
```


<h1 style="color:green;font-size:19px">Построить графики п.2 и 3 в одной координатной плоскости, сделать подписи графиков, указать легенду.</h1>


```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-10, 10, 100)

y1 = x ** 2 + 3 * x + 2
y2 = 2 * np.cos(5 * x + 3)

fig, ax1 = plt.subplots(figsize=(8, 6))

color = 'cyan'
ax1.set_xlabel('X')
ax1.set_ylabel('y = x^3 - 2x^2 + x + 6', color=color)
ax1.plot(x, y1, label='y = x^3 - 2x^2 + x + 6', color=color, linewidth=2)
ax1.tick_params(axis='y', labelcolor=color)

ax2 = ax1.twinx()

ax1.spines['left'].set_position('center')
ax1.spines['bottom'].set_position('zero')
ax1.spines['right'].set_color('none')
ax1.spines['top'].set_color('none')

color = 'r'
ax2.set_ylabel('y = 2 * cos(5x + 3)', color=color)
ax2.plot(x, y2, label='y = 2 * cos(5x + 3)', color=color, linewidth=2)
ax2.tick_params(axis='y', labelcolor=color)

plt.title('Графики функций')
plt.grid(True)

lines, labels = ax1.get_legend_handles_labels()
lines2, labels2 = ax2.get_legend_handles_labels()
ax2.legend(lines + lines2, labels + labels2, loc='best')

plt.show()
```

<h1 style="color:red;font-size:19px">${x(t)} = \cos(3t)$, $y(t) = \sin(2t)$, $\(t \in [0, 4\pi]\)$</h1>



```python
t = np.linspace(0, 4 * np.pi, 1000)

x = np.cos(3 * t)
y = np.sin(2 * t)

plt.figure(figsize=(8, 6))
plt.plot(x, y, label='x(t) = cos(3t), y(t) = sin(2t)', markerfacecolor='blue', marker='o', color='r', linewidth=2)

plt.xlabel('x(t)')
plt.ylabel('y(t)')
plt.title('График параметрической функции')

plt.legend()

plt.grid(True)
plt.show()
```



<h1 style="color:green;font-size:19px">Построить столбиковую диаграмму по данным статистического комитета КР по прожиточному минимуму, представленному в таблице</h1>

```python
data = np.array([["Наименование показателей", 2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019, 2020, 2021, 2022, ],
             ["Кыргызская Республика", 4341.5, 4599.21, 4981.51, 5182.99, 4794.34, 4900.79, 4792.54, 4806.32, 5358.53,
              6268.31, 7178.32 ]])
data
```

```python
years = data[0, 1:]
values = data[1, 1:]

plt.figure(figsize=(10, 6))
plt.bar(years, values, color='g', alpha=0.7)

plt.xlabel('Год')
plt.ylabel('Прожиточный минимум')
plt.title('Прожиточный минимум в Кыргызской Республике (2012-2022)')

plt.xticks(rotation=45)

plt.grid(axis='y')
plt.tight_layout()  
plt.show()
```


<h1 style="color:green;font-size:19px">Построить круговой график по данным п. 7</h1>

```python
years = data[0, 1:]
values = data[1, 1:]

max_value_index = np.argmax(values)

explode = [0.1 if i == max_value_index else 0 for i in range(len(years))]

plt.figure(figsize=(8, 8))
plt.pie(values, labels=years, autopct='%1.1f%%', startangle=140, colors=plt.cm.Paired.colors, explode=explode)

plt.title('Прожиточный минимум в Кыргызской Республике (2012-2022)')

plt.axis('equal')
plt.show()
```


<h1 style="color:green;font-size:19px">Построить график произвольной каркасной поверхности.</h1>

```python
import numpy as np
import matplotlib.pyplot as plt

R = 1
r = 0.3

u = np.linspace(0, 2 * np.pi, 100)
v = np.linspace(0, 2 * np.pi, 100)
u, v = np.meshgrid(u, v)

x = (R + r * np.cos(v)) * np.cos(u)
y = (R + r * np.cos(v)) * np.sin(u)
z = r * np.sin(v)

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(x, y, z, cmap='viridis')

ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title('Тор (Донут)')

plt.show()
```
