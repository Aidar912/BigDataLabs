```python
#Импорты
import numpy as np
from stop_words import get_stop_words
```

#### Создайте свой «мешок слов», например, набор из 20-25 отзывов о чем-либо, напишите краткое пояснение к данным.

```python
#Данные содержат коллекцию отзывов об отеле, включающую в себя как положительные, так и отрицательные мнения по поводу различных аспектов пребывания в отеле.
reviews = [
    "Прекрасный отель! Все номера очень чистые и уютные, персонал приветливый и отзывчивый.",
    "Отличное обслуживание, прекрасный вид из номеров на море.",
    "Великолепное расположение, рядом с пляжем, множество развлечений вокруг.",
    "Замечательный завтрак! Большой выбор блюд, всегда свежие продукты.",
    "Уютный отель с прекрасной атмосферой и отличным сервисом.",
    "Превосходный персонал, быстро реагирующий на все запросы гостей.",
    "Номера убраны и оборудованы всем необходимым, просторные и комфортные кровати.",
    "Чистый и ухоженный бассейн, идеальное место для отдыха.",
    "Восхитительный вид с балкона номера на море, особенно на закате.",
    "Отель, в котором всегда приятно останавливаться, отличное соотношение цены и качества.",
    "Ужасное обслуживание! Персонал невнимательный и неотзывчивый.",
    "Номера грязные и запущенные, не соответствуют заявленному уровню.",
    "Плохая звукоизоляция, слышно все звуки с улицы и из соседних номеров.",
    "Завтраки не впечатляют, маленький выбор и не всегда свежие продукты.",
    "Неудобные кровати, спать на них не комфортно.",
    "Нет парковки, приходится долго искать место для автомобиля.",
    "Отсутствие кондиционера в номерах, особенно это мешает в жаркое время года.",
    "Не работающий Wi-Fi, сигнал очень слабый.",
    "Недружелюбный персонал на ресепшене, не умеют решать проблемы гостей.",
    "Отель не соответствует ожиданиям, цена завышена для такого уровня услуг.",
    "Плохое обслуживание, персонал не всегда готов помочь.",
    "Номера маленькие и неудобные для проживания.",
    "Неудачное местоположение отеля, далеко от основных достопримечательностей.",
    "Несвежая мебель и интерьер в номерах, требует обновления.",
    "Очень дорогой отель для таких услуг, не стоит своих денег.",
    "Слабая звукоизоляция между номерами, слышно все разговоры соседей.",
    "Уборка номеров проводится плохо, не меняют постельное белье и полотенца.",
    "Плохое освещение в номерах, темно даже днем.",
    "Отсутствие удобств, которые обещали при бронировании.",
    "Отель старый и запущенный, требует ремонта и обновления.",
]
labels = np.array([1]*15 + [0]*15)
```

```python
print(len(reviews))
print(len(labels))
```

#### Создайте словарь стоп-слов

```python
# Получение стоп-слов для русского языка с помощью библиотеки stop-words
stop_words = get_stop_words('russian')

# Вывод полученных стоп-слов
print(stop_words)
```

#### Примените стоп-слова к своему «мешку слов»

```python
# Очистка списка от стоп-слов на основе русского языка
# Удаление стоп-слов из отзывов, чтобы получить "мешок слов" без них

bag_of_words_without_stopwords = []

for review in reviews:
    words = review.split()
    filtered_words = [word for word in words if word.lower() not in stop_words]
    bag_of_words_without_stopwords.append(' '.join(filtered_words))

# Вывод "мешка слов" без стоп-слов
print("Мешок слов без стоп-слов:")
for review in bag_of_words_without_stopwords:
    print(review)
```

#### Масштабируйте данные с помощью tf-idf

```python
from sklearn.feature_extraction.text import TfidfVectorizer

tfidf_vectorizer = TfidfVectorizer()

tfidf_matrix = tfidf_vectorizer.fit_transform(bag_of_words_without_stopwords)

print("TF-IDF матрица:")
print(tfidf_matrix.toarray())


```

```python
feature_names = tfidf_vectorizer.get_feature_names_out()  # Получение списка признаков (слов)

idf_values = tfidf_vectorizer.idf_

words_idf = sorted(zip(feature_names, idf_values), key=lambda x: x[1], reverse=True)

for word, idf in words_idf:
    print(f"{word}: {idf:.4f}")
```

#### Исследуйте коэффициенты модели

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression()

model.fit(tfidf_matrix, labels)

feature_names = tfidf_vectorizer.get_feature_names_out()
coefficients = model.coef_[0]

for word, coef in sorted(zip(feature_names, coefficients), key=lambda x: x[1], reverse=True):
    print(f"{word}: {coef:.4f}")
```

```python
import matplotlib.pyplot as plt

words, idf_values = zip(*words_idf[:20])

plt.figure(figsize=(10, 6))
plt.barh(words, idf_values)
plt.xlabel('Значения IDF')
plt.title('Топ-20 слов по значениям IDF')
plt.gca().invert_yaxis()
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

tfidf_array = tfidf_matrix.toarray()

plt.figure(figsize=(10, 6))
sns.heatmap(tfidf_array, cmap='viridis', xticklabels=feature_names, cbar=True)
plt.title('Визуализация матрицы TF-IDF')
plt.xlabel('Слова')
plt.ylabel('Отзывы')
plt.show()
```

```python
from sklearn.manifold import TSNE

perplexity_value = 10

tsne = TSNE(random_state=42, perplexity=perplexity_value)
tfidf_embedded = tsne.fit_transform(tfidf_matrix.toarray())

positive_indices = [i for i, label in enumerate(labels) if label == 1]
negative_indices = [i for i, label in enumerate(labels) if label == 0]

plt.figure(figsize=(10, 8))
plt.scatter(tfidf_embedded[positive_indices, 0], tfidf_embedded[positive_indices, 1], label='Положительные отзывы', alpha=0.5)
plt.scatter(tfidf_embedded[negative_indices, 0], tfidf_embedded[negative_indices, 1], label='Отрицательные отзывы', alpha=0.5)

plt.title('t-SNE Визуализация Положительных и Отрицательных Отзывов')
plt.xlabel('t-SNE особенность 1')
plt.ylabel('t-SNE особенность 2')
plt.legend()
plt.show()
```

Коэффициенты, полученные из модели, позволяют оценить важность определенных слов и фраз в отзывах о ресторанах. Некоторые слова, такие как "прекрасный", "отличное", "великолепное" и "уютный", связаны с положительными оценками и высоким качеством обслуживания или услуг. С другой стороны, слова с отрицательными коэффициентами, такие как "плохое", "недружелюбный", "требует" и "неудачное", указывают на проблемы или негативный опыт клиентов.
Слова с низкими значениями коэффициентов, такие как "белье", "интерьер", "соседей", "денег", играют менее значимую роль в общем восприятии отзыва, несмотря на их наличие.

Этот анализ помогает понять, какие аспекты оцениваются клиентами выше и на что стоит обратить внимание для улучшения качества предоставляемых услуг. Следует уделить особое внимание аспектам, связанным с положительными оценками, чтобы повысить удовлетворенность клиентов и общую репутацию ресторана.

Ключевые аспекты положительных отзывов: Описание удобств, окружающей среды, услуг и обслуживания. Слова, связанные с удобствами или качеством услуг, имеют сильное положительное влияние.

Ключевые аспекты отрицательных отзывов: Отзывы сконцентрированы на негативных аспектах, таких как низкое качество, требующие улучшений услуги или предоставляемые условия.
